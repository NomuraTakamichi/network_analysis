```python
from konlpy.tag import Hannanum
import csv
import numpy as np
from requests_oauthlib import OAuth1Session
import json
import tweepy
```

```python
hannanum = Hannanum()

text_list = []
photo_list = []
user_name = []
text_name = []
photo_name = []

#with open('./data/bts.csv','r', encoding='utf-8') as file:
with open('./data/netflix/퀸스갬빗.csv','r', encoding='utf-8')as file:
    reader = csv.reader(file)
    for i,row in enumerate(reader):
        tweet_check = 0
        name_check = 0
        if i % 2 == 0:
            text = hannanum.pos(row[1])
            
            for j in range(len(text)):
                if text[j][1] == "F":
                    tweet_check = 1
                    break
                elif text[j][1] == "N":
                    #print(text)
                    tweet_check = 2
                else:
                    tweet_check = 0
            #user중복 방지
            for name in user_name:
                if name == row[0]:
                    name_check = 1
                    break
            if name_check == 0:
                if tweet_check == 0: 
                    text_list.append(row[1])
                    text_name.append(row[0])
                elif tweet_check == 1:
                    photo_list.append(row[1])
                    photo_name.append(row[0])
                user_name.append(row[0])

# text_list : text로만 되어 있는 tweet의 수 
print("text_list:",len(text_list))
# photo_list : photo나 movie를 보함한 tweet의 수 
print("photo_list:",len(photo_list))
print(len(text_name))
print(len(photo_name))
```

```python
url = "https://api.twitter.com/1.1/statuses/user_timeline.json"

text_timeline_list = []
for i,name in enumerate(text_name):
    #print("================================================")
    #print(name)
    # パラメータ
    params = {
        "count": 50,
        "screen_name": name,
        "include_entities": True,
        "exclude_replies": False,
        "include_rts": False,
    }
    # response로부터 timeline list취득
    res = twitter.get(url, params=params)
    if res.status_code == 200:
        text = json.loads(res.text)
        for line in text:
            text_timeline_list.append(hannanum.pos(line['text']))
    text_timeline_list.append([ ])
```
